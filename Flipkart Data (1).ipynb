```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
# 1️⃣ Load data


df = pd.read_csv(r"D:\archive (2)\flipkart_com-ecommerce_sample.csv")
df
```

```python
df.columns
```

```python
# 2️⃣ Check first 5 rows


df.head()
```

```python
# 3️⃣ Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
# 4️⃣ Handle missing values
df['retail_price'] = df['retail_price'].fillna(0)
df['discounted_price'] = df['discounted_price'].fillna(0)
df['overall_rating'] = df['overall_rating'].fillna(0)
```

```python
# 5️⃣ Convert prices to numeric (if they are not)
df['retail_price'] = pd.to_numeric(df['retail_price'], errors='coerce')
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

```

```python
# 6️⃣ Calculate discount percentage
df['discount_percent'] = ((df['retail_price'] - df['discounted_price']) / df['retail_price'] * 100).round(2)
```

```python
print(df[['product_name','brand','retail_price','discounted_price','discount_percent','overall_rating']].head())
```

```python
# 8️⃣ Top 10 brands by number of products
top_brands = df['brand'].value_counts().head(10)
print(top_brands)
```

```python
# 9️⃣ Plot top brands
top_brands.plot(kind='bar', figsize=(10,5), title='Top 10 Brands by Number of Products')
plt.show()
```

```python
# 🔟 Plot histogram of discounted prices
plt.figure(figsize=(10,5))
plt.hist(df['discounted_price'], bins=50, color='skyblue')
plt.title('Distribution of Discounted Prices')
plt.xlabel('Discounted Price')
plt.ylabel('Number of Products')
plt.show()
```

```python
# 11️⃣ Save cleaned dataset for Power BI
df.to_csv("D:/Flipkart data/flipkart_cleaned.csv", index=False)
```
